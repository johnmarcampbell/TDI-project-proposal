```python
import pandas as pd
import pickle
import matplotlib.pyplot as plt


monthly = pd.read_pickle('monthly.pkl')
```

```python
labels = []
for index, _ in monthly.iterrows():
    labels.append(int(index[0]) + int(index[1]) / 12.)
    
fig, ax = plt.subplots()
plt.scatter(labels,[1e4 * x for x in monthly['gun ratio'].tolist()])
plt.axis([2009,2019,-0.9,20])
```

```python
fig, ax = plt.subplots()
plt.scatter(labels,[1e4 * x for x in monthly['russia ratio'].tolist()])
plt.axis([2009,2019,-.9,7])
```
